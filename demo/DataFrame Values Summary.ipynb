```python
from sklearn.datasets import load_diabetes
```

```python
import sumnplot as sp
from sumnplot.discretisation import EqualWidthDiscretiser
from sumnplot.summary import DataFrameValueCounter
```

```python
sp.__version__
```

# Load data
Add a categorical version of `bmi` onto the dataset.

```python
X, y = load_diabetes(return_X_y=True, as_frame=True)
```

```python
X["bmi_bucketed"] = EqualWidthDiscretiser(variable="bmi").fit_transform(X)
```

```python
X.head()
```

# Summarise values in DataFrame

```python
df_summariser = DataFrameValueCounter(
    columns=["age", "sex", "bmi_bucketed", "bp"], max_values=20, summary_values=5
)
```

```python
df_summariser.summarise(X)
```
